```python
import matplotlib.pyplot as plt
import scipy
from scipy.interpolate import interp1d
import numpy as np
```

```python
def interpolate(x, y, new_x):
    """
    Интерполировать набор точек (x, y) и получить новый набор значений Y
    по новому набору X.
    """
    fn = interp1d(x, y, kind='cubic')
    return fn(new_x)
```
